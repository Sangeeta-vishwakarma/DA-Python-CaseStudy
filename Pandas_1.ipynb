**11. Add new column.**

```python
# 1. Approach
df['Half_Marks'] = df['Marks']/2  # new column having half marks of 'marks' column
df
```

```python
#2. Approach
# create function and use apply method to 'half marks'
def half_makrs(x):
  return x/2

df['Half_Marks'] = df['Marks'].apply(half_makrs)
df
```

```python
# 3. Approach
# use apply method on lambda function
df['half marks'] = df['Marks'].apply(lambda x:x//2)
df
```

**12. Map Function**
